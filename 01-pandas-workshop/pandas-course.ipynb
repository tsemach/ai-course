```python
# iterate over rows
for rowIndex, row in df_new.iterrows():
    for columnIndex, value in row.items():
        if rowIndex < 5:
            print(value)
```

```python
# Apply with lambda function
df_new['m_per_hour'] = df_new['m_per_sec'].apply(lambda x: x * 60*60)
df_new.head()
```

```python
# Define a new function
def extract_point_location(x):
    return x.split('_')[-1]
```

```python
# Apply with your function
df_new['point_location'] = df_new['pointId'].apply(extract_point_location)
df_new.head()
```
